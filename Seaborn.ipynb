```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
sns.countplot(x = 'sex',data = df)
```

```python
sns.countplot(y = 'sex',data = df)
```

```python
sns.countplot(x = 'smoker',data = df)
```

```python
sns.barplot(x = 'total_bill',y = 'sex',data = df)
```

## Box Plot 

```python
sns.boxplot('smoker','total_bill',data = df)
```

```python
sns.boxplot('sex','total_bill',data = df, palette = 'rainbow')
```

```python
sns.boxplot(data = df,orient = 'v')
```

```python
sns.boxplot(x ='total_bill', y = 'smoker',hue = 'sex',data= df)
```

## Violin Plot

```python
sns.violinplot(x='total_bill',y = 'day',hue = 'sex',data= df,palette = 'rainbow')
```
