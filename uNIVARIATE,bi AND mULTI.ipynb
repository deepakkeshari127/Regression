### Univariate Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df.head()
```

```python
setosa_df = df.loc[df['species']=='setosa']
```

```python
virginica_df  = df.loc[df['species']=='virginica']
```

```python
df['species'][70]
```

```python
versicolor_df = df.loc[df['species']=='versicolor']
```

```python
df.iloc[2:75].head()
```

```python
plt.plot(setosa_df['sepal_length'],np.zeros_like(setosa_df['sepal_length']))
plt.plot(virginica_df['sepal_length'],np.zeros_like(virginica_df['sepal_length']))
plt.plot(versicolor_df['sepal_length'],np.zeros_like(versicolor_df['sepal_length']))
```

# Bivariate 

```python
sns.FacetGrid(df, hue = "species",size = 5).map(plt.scatter, "sepal_length", "sepal_width", edgecolor="w");
plt.legend()
plt.show()
```

```python
sns.FacetGrid(df, hue = "species",size = 5).map(plt.scatter, "sepal_length", "sepal_width", edgecolor="w").add_legend()
plt.show()
```

```python
g = sns.FacetGrid(df, col = "species",size = 2).map(plt.scatter, "sepal_length", "sepal_width", edgecolor="w")
```
